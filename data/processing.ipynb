```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
valid = df.loc[df['player.age'].notnull()]
valid = valid.loc[valid['player.age'] != 99]
valid = valid.loc[valid['player.age'] != 2]
print(len(valid))
valid.iloc[0]
```

```python
len(valid.loc[valid['participant._is_bot'] != 0])
```

```python
cols = [
    'player.queries_issued', 
    'player.age', 
    'player.gender', 
    'player.profession', 
    'player.total_task_time', 
    'player.treatment',
    'player.topicname'
]
valid.loc[:, cols].sort_values(by=['player.age'])
```

```python
print("Mean age: {}".format(np.mean(valid['player.age'])))
print("Std dev age: {}".format(np.std(valid['player.age'])))
print("Male/female/other: {}/{}/{}".format(
    len(valid.loc[valid['player.gender'] == "male"]),
    len(valid.loc[valid['player.gender'] == "female"]),
    len(valid.loc[valid['player.gender'] == "other"])
))
print("student/working/uni_staff/none: {}/{}/{}/{}".format(
    len(valid.loc[valid['player.profession'] == "student"]),
    len(valid.loc[valid['player.profession'] == "working"]),
    len(valid.loc[valid['player.profession'] == "uni_staff"]),
    len(valid.loc[valid['player.profession'] == "none"]),
))
print(valid.groupby('player.topicname')['participant.code'].nunique())
```

```python
valid.to_csv('filtered_data.csv')
```
